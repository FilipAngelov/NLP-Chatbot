# Provide and prepare data information

```python
questions_array = questions.to_numpy()
questions_array
```

```python
stop_words_mkd = pd.read_csv('stop_words.txt').to_numpy()
stop_words_list = []
for words in stop_words_mkd:
    for word in words:
        stop_words_list.append(word)
        
stop_words_list
```

```python
questions_vectorizer = CountVectorizer(stop_words=stop_words_list)
```

```python
questions_vectorizer
```

```python
questions_feature = questions_vectorizer.fit_transform(questions_array)
questions_df = pd.DataFrame(data = questions_feature.todense(), 
                                    columns=questions_vectorizer.get_feature_names())

questions_df.tail(10)
```
